```python
import pandas as pd
import nltk

from nltk.corpus import stopwords
stop_words=stopwords.words('indonesian')
stop_words.extend(["yg", "dg", "dgn", "ny", "d", 'klo', 
                       'kalo', 'amp', 'biar', 'bikin', 'bilang', 
                       'gak', 'ga', 'krn', 'nya', 'nih', 'sih', 
                       'si', 'tau', 'tdk', 'tuh', 'utk', 'ya', 
                       'jd', 'jgn', 'sdh', 'aja', 'n', 't', 
                       'nyg', 'hehe', 'pen', 'u', 'nan', 'loh',
                       '&amp', 'yah'])

df = pd.read_csv('SW_ShopeeFood.csv', names=['stopwords']) 
stop_words.extend(df['stopwords'][0].split(' '))
stop_words = set(stop_words)

def stopwords_removal(words):
    return [word for word in words if word not in stop_words]

df['tweet_tokens_WSW'] = df['Tokenizing'].apply(stopwords_removal) 
```
