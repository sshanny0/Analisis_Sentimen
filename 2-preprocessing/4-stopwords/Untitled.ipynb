```python
import nltk

from nltk.corpus import stopwords
```

```python
with open('SW_Only_ShopeeFood.csv', 'r', encoding='utf-8') as f:
    stop_words = f.read().strip().split(" ")
stop_words
```

```python
words = set(stopwords.words("indonesian"))
filtered = [w for w in stop_words if not w in words]
```

```python
filtered
```

```python
list_stopwords = stopwords.words('indonesian')

txt_stopword = pd.read_csv("SW_ShopeeFood.csv", names= ["Data_Cleaning"], header = None)

# convert stopword string to list & append additional stopword
list_stopwords.extend(txt_stopword["Data_Cleaning"][0].split(' '))

list_stopwords = set(list_stopwords)
```

```python
list_stopwords
```

```python
def stopwords_removal(words):
    return [word for word in words if word not in list_stopwords]

tweet_df['StopWords'] = tweet_df['Data_Cleaning'].apply(stopwords_removal) 


tweet_df
```
