```python
import pandas as pd 
import numpy as np
import string
```

```python
datanya = pd.read_csv('crawling-semua.csv')

datanya
```

```python
df  = pd.DataFrame(datanya[['Tweets']])
```

```python
def case_fold(tweet):
    tweet = tweet.lower()
    return tweet

df['Case_Folding'] = df['Tweets'].apply(case_fold)
```
