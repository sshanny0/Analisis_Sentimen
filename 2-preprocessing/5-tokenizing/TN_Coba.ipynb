```python
import pandas as pd 
import numpy as np
import string 
import re #regex library
import nltk
```

```python
from nltk.tokenize import word_tokenize 
```

```python
def load_data():
    data = pd.read_csv('TN_ShopeeFood.csv')
    return data
```

```python
tweet_df = load_data()
tweet_df.head(100)
```

```python
df  = pd.DataFrame(tweet_df[['Tweets', 'Case_Folding', 'Data_Cleaning']])
```

```python
def clean_tweets(tweet):
    
    # tokenize tweets
    tokens = nltk.tokenize.word_tokenize(tweet)
    
    return tweet
```

```python
df['Tokenizing'] = df['Data_Cleaning'].apply(clean_tweets)
```

```python
df.to_csv('TN_ShopeeFood.csv',encoding='utf8', index=False)
```
