```python
normalizad_word = pd.read_csv('Kamus_Baku.csv')

normalizad_word_dict = {}

for index, row in normalizad_word.iterrows():
    if row[0] not in normalizad_word_dict:
        normalizad_word_dict[row[0]] = row[1] 

def normalized_term(document):
    return ' '.join([normalizad_word_dict[term] if term in normalizad_word_dict else term for term in document.split()])

df['Tweet_Normal'] = df['Clean_Data'].apply(normalized_term)

df
```

```python
df.to_csv('normal-semua.csv',encoding='utf8', index=False)
```

```python
spell = pd.read_excel('kamus.xlsx')

# make a place where we can build our new sentence
new_sentence = []

# look through each word 
for word in spell:
    # look for each candidate
    for candidate_replacement in spell:
        # if our candidate is there in the word
        if candidate_replacement in word:
            # replace it 
            word = word.replace(candidate_replacement, spell[candidate_replacement])

    # and pop it onto a new list 
    new_sentence.append(word)

rfrm = " ".join(new_sentence)
print(rfrm)
```
