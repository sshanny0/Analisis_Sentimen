```python
import pandas as pd
import nltk
import numpy as np
import Sastrawi

from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def load_data():
    data = pd.read_csv('sw-semua.csv')
    return data

tweet_df = load_data()
tweet_df
```

```python
df = pd.DataFrame(tweet_df[['Tweets', 'Case_Folding', 'Clean_Data', 'Tweet_Normal', 'Tanpa_StopWords']])
```

```python
# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

df['Stemming'] = df['Tanpa_StopWords'].apply(lambda x : ([stemmer.stem(y) for y in x.split()]))
```

```python
df
```

```python
df.to_csv('stemming-semua.csv',encoding='utf8', index=False)
```
